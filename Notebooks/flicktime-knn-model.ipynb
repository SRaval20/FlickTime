```python
def get_genre_ids(genre_list):
    genre_ids = []
    for genre in genre_list:
        genre_ids.append(genre['id'])
    return genre_ids

new_moviedf['genre_id'] = new_moviedf['genres'].apply(get_genre_ids)
```

```python
# Define a function to encode movie features using autoencoders
def encode_movies(movies):
    inputs = Input(shape=(len(movies.columns),))
    encoded = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(inputs)
    encoded = Dense(64, activation='relu', kernel_regularizer=l2(0.001))(encoded)
    encoded = Dense(32, activation='relu', kernel_regularizer=l2(0.001))(encoded)
    model = Model(inputs, encoded)
    optimizer = Adam(lr=0.001)
    model.compile(optimizer=optimizer, loss='mse')
    early_stop = EarlyStopping(monitor='val_loss', patience=10)
    model.fit(movies, movies, batch_size=32, epochs=100, validation_split=0.2, callbacks=[early_stop])
    encoded_movies = model.predict(movies)
    return encoded_movies
```

# K Nearest Neighbors Algorithms

```python
%%time
from sklearn.decomposition import PCA

pca = PCA(n_components=200)
X_pca = pca.fit_transform(X)
```

```python
knn = NearestNeighbors(n_neighbors=50, algorithm='ball_tree')
knn.fit(X_pca)
```

```python
def get_recommendations(movie_title):
    movie_index = train_data[train_data['title'] == movie_title].index[0]
    distances, indices = knn.kneighbors(X_pca[movie_index].reshape(1,-1))
    recommended_movies = []
    for i in range(1, len(indices[0])):
        recommended_movies.append(train_data.iloc[indices[0][i]]['title'])
    return recommended_movies
```
